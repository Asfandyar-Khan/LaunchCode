```python
# -----------------------------
# Daily Quiz Challenge üéØ
# -----------------------------

# Dictionary = stores questions and their correct answers
# The "question" is the KEY and the "answer" is the VALUE
quiz = {
    "What is 5 + 3?": "8",
    "What color is the sky on a clear day?": "blue",
    "What is the capital of France?": "paris"
}

# List to remember which questions were answered correctly already
answered_correctly = []

# Keep track of the player's score
points = 0


# -----------------------------
# FUNCTIONS (game actions)
# -----------------------------

def ask_question(question):
    """Ask the player a question and check their answer"""
    global points  # allows us to update the points variable from outside this function

    # Show the question
    print("\n" + question)

    # Ask the player for their answer
    # .lower() makes it lowercase so "Blue" and "blue" both work
    answer = input("Your answer: ").lower()

    # Check if the answer is correct
    if answer == quiz[question].lower():
        print("‚úÖ Correct!")
        points += 1   # add a point to score
        answered_correctly.append(question)  # mark this question as completed
    else:
        print("‚ùå Oops! The correct answer was:", quiz[question])


def start_game():
    """This runs the quiz game"""
    print("üéØ Welcome to the Daily Quiz Challenge!")
    print("Answer all the questions to win!\n")

    global points
    points = 0   # reset score to 0 each new game
    answered_correctly.clear()  # clear old answers

    # Keep asking questions until ALL are answered correctly
    while len(answered_correctly) < len(quiz):
        for question in quiz:
            if question not in answered_correctly:  # only ask if not answered yet
                ask_question(question)

    # Once all questions are done, show results
    print("\nüéâ You answered all questions correctly!")
    print("Your total points:", points)

    # Ask if the player wants to play again
    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again == "yes":
        start_game()  # restart the game (recursion!)
    else:
        print("Thanks for playing! üëã")


# -----------------------------
# START THE GAME
# -----------------------------
start_game()
```
